# Limpieza y Procesamiento de Datos Climáticos de Bogotá
## Fuente: Open-Meteo Historical Weather API
### Período: 2024-01-01 al 2024-12-31

```python
#Celda - 2 importar librerias 

#Librerias necesarias para la limpieza y procesamiento
import pandas as pd  #para manipulacion de datos 
import json          #para leer el archivo JSON crudo
import os            #para manejar ruta de archivos
```

```python
#Celda 3 - Cargar el JSON desde data/raw
ruta_json = os.path.join("..", "data", "raw", "clima_bogota_2024.json")

with open(ruta_json, "r", encoding="utf-8") as f:
    data = json.load(f)

print(type(data))
print(data.keys())
```

```python
#Celda 4
#explorar la estructura de los datos horarios 
print(type(data["hourly"]))
print(data["hourly"].keys())
```

```python
#Celda 5 - convertir los datos horariuos en un DataFrame
df = pd.DataFrame(data["hourly"])

#verifivar las primeras filas
print(df.shape)
df.head()
```

```python
#celda 6 
#revisar ripos de datos y valores nulos 

print(df.dtypes)
print()
print(df.isnull().sum())
```

```python
#celda 7 - convertir la columna time a formato datetime

df["time"] = pd.to_datetime(df["time"])

#verificar el cambio
print(df.dtypes)
print()
print(df["time"].head())
```

```python
#celda  8 - establecer time como indice del dataframe

df = df.set_index("time")

#verirficar resultado
df.head
```

```python
#celda 9 - guardar el DataFrame limpio en data/processed/
ruta_processed = os.path.join("..", "data", "processed", "clima_bogota_limpio.csv")

df.to_csv(ruta_processed)

print("achivo guardado correctamente")
```
