# Extracción de Datos Climáticos de Bogotá
## Fuente: Open-Meteo Historical Weather API
### Período: 2024-01-01 al 2024-12-31

```python
#Celda 1 - importas librerias
import requests     #para hacer llamadas http a la API
import json         #Para guardar la respuesta en formato JSON
import os           #para manejar rutas de archivo y carpetas
```

```python
#Celda 2 - parametros de la consulta
#parametros de la consulta a la API de Open-Meteo
#Cordenadas de Bogota
LATITUDE = 4.71
LONGITUDE = -74.07

#Rango de fechas hitoricas a consultar 
START_DATE = "2024-01-01"
END_DATE = "2024-12-31"

#Variables climaticas de interes
VARIABLES = "temperature_2m","precipitation","windspeed_10m","relativehumidity_2m"
```

```python
#Celda 3 - Construir la URL y hacer la llamada
#construir la URL  con los parametros definidos
BASE_URL = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": LATITUDE,
    "longitude": LONGITUDE,
    "start_date": START_DATE,
    "end_date":  END_DATE,
    "hourly":    VARIABLES
}

#hacer la llamada a la API
response = requests.get(BASE_URL, params=params)
```

```python
#celda 4 - verifica la respuesta
#verificar que la API respondio correctamente 
#el codigo 200 significara exito 
print(response.status_code)

print(response.json())
print(response.url)
```

```python
#Clda 5 - extraer el JSON y guardarlo en data/raw
#extraer el contenido JSON de la respuesta
data = response.json()

#definir la ruta donde se guardara el archivo crudo
output_path = os.path.join("..", "data","raw", "clima_bogota_2024.json")

#guardar el JSON en disco
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("archivo guardado correctamente")
```
